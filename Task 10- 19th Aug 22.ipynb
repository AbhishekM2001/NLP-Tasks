```python
#Loading the dataset
import pandas as pd
df=pd.read_csv("IMDB Dataset.csv")
```

```python
df.head()
```

```python
#Data Preprocessing - Remove the HTML tags
```

```python
from nltk.stem.porter import PorterStemmer
porter=PorterStemmer()

def stemmer(text):
    return [porter.stem(word) for word in text.split()]
```

```python
#Vectorization of Documents
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf= TfidfVectorizer(strip_accents=None,lowercase=False,tokenizer=stemmer,norm='l2',use_idf=True)
```

```python
Y=df.sentiment.values
X=tfidf.fit_transform(df.review)
```

```python
# Document Classification using Logistic Regression

from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test=train_test_split(X,Y,random_state=1,test_size=0.5,shuffle=False)

import pickle

from sklearn.linear_model import LogisticRegressionCV
logit=LogisticRegressionCV(cv=5,scoring='accuracy',max_iter=100).fit(X_train,Y_train)
saved_model=open('saved_model.sav','wb')
pickle.dump(logit,saved_model)
saved_model.close()
```

```python
# Evaluating the model
filename='saved_model.sav'
saved_logit=pickle.load(open(filename,'rb'))

saved_logit.score(X_test,Y_test)
```
